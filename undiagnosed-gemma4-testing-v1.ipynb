```python
import gc

def offload_model_from_memory():
    # delete the models and tokenizers
    if 'model' in locals():
        del model
    if 'tokenizer' in locals():
        del tokenizer

    # force run the garbage collector
    gc.collect()
    
    # empty torch cuda and release memory back to both gpus
    with torch.no_grad():
        torch.cuda.empty_cache()
```

**Note:** Right now we will first work with the Gemma 4 E4B since using the 26B model will require sharding and more powerful accelerators, and JAX and Keras. Therefore, the entire testing phases and architecture will be built using the E4B and will later be swapped with the 26B MoE model.

```python
# load gemma 4
from transformers import AutoTokenizer, AutoModelForMultimodalLM, pipeline, BitsAndBytesConfig, AutoProcessor
import torch

"""
The model needs to load all its weights in both GPUs.
This can lead to a problem later when all the tensors will be expected to be on the 
same device. Therefore, 4-bit quantization will allow for all the model parameters
to be loaded onto the same GPUs.
"""
# this is essentially needed for the 26B model; the E4B does not require quantization to be fully loaded into the 2xT4 GPUs
bnb_config = BitsAndBytesConfig(
    load_in_4bit = True,
    bnb_4bit_compute_dtype=torch.bfloat16, # the same as specified in the model dtype 
    bnb_4bit_use_double_quant=True, # to same some more VRAM
    bnb_4bit_quant_type="nf4",
    llm_int8_enable_fp32_cpu_offload=True
)

# taken from the preloaded model
GEMMA4_E4B_MODEL_ID = "/kaggle/input/models/google/gemma-4/transformers/gemma-4-e4b-it/1"
GEMMA4_26B_MODEL_ID = "/kaggle/input/models/google/gemma-4/transformers/gemma-4-26b-a4b-it/1"

# for tokenizing, encoding, and decoding, and managing the tokens
tokenizer = AutoTokenizer.from_pretrained(GEMMA4_E4B_MODEL_ID)

processor = AutoProcessor.from_pretrained(GEMMA4_E4B_MODEL_ID)

# model configuration
model = AutoModelForMultimodalLM.from_pretrained(
    GEMMA4_E4B_MODEL_ID,
    # quantization_config=bnb_config,
    dtype = torch.bfloat16,
    device_map = "auto", # automatically splits across the accelerators
    low_cpu_mem_usage = True # prevents 'meta' device issue by allocating memory before loading
)

print(f"Model loaded on: {model.hf_device_map}")
print(f"GPU memory allocated: {torch.cuda.memory_allocated() / 1e9: .1f} GB")
```

```python
def ask_gemma_test(prompt, max_new_tokens=512):
    # formatting the prompt according to Gemma 4's chat template
    messages = [
        {
            "role": "system",
            "content": "You are a Clinician-Scientist working at a world-class research institution that aims to democratize the awareness of cardiac-related diseases. You specialise more into explanining the ambiguity behind vague descriptions of symptoms that someone might have that they think might lead to a cardiac issue, and help them understand what the symptoms actually entail in reality."
        },
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": prompt
                }
            ]
        }
    ]

    # apply the chat template to the messages
    formatted = processor.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
        enable_thinking=True # triggers the <|think|> logic
    )

    # process the inputs
    inputs = processor(
        text=formatted,
        return_tensors="pt", 
    ).to(model.device) # send it to the model's accelerators, where the embedding tokens live (embed_tokens)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens = max_new_tokens,
            do_sample = True,
            # low temperatures allow for generations to be more precise and deterministic,
            # higher temperatures are often recommended for thinking models
            temperature = 0.7, 
            top_p = 0.95
        )

    # keep the special tokens to see the reasoning tags
    response = processor.decode(
        outputs[0][inputs['input_ids'].shape[1]:],
        skip_special_tokens=False
    )

    return response
```
